```python
def movie_scrap_id(movie_name):
    try:
        headers={"User-Agent" : "Mozilla/5.0"}
        url = f'https://www.imdb.com/find/?q={movie_name.replace(" ","+")}' # In URLs, spaces are not allowed.
  # + Commonly used in query strings like search
        response = requests.get(url, headers=headers)
        soup=BeautifulSoup(response.text,"html.parser")
    #print(soup)
        imdb_id = None
        for A_tag in soup.find_all('a', href=True):
            href = A_tag['href']
            if href.startswith('/title/tt'):
                imdb_id = href.split('/')[2]
                    #print("First IMDb ID:", imdb_id)
                break

        return imdb_id


    except Exception as e:
        print(e)
        return None
```
