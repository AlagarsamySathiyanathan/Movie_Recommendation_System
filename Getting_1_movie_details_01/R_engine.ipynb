```python
# importing libraries
import requests 
from bs4 import BeautifulSoup
import openpyxl
import pandas as pd
```

## try-except 
# In Python, try-except (commonly known as try-catch in other languages) is used for exception handling — to catch and handle errors during program execution.
# It’s a way to run code that may produce an error and handle the error gracefully without crashing your program.


```python

try:
    headers={"User-Agent" : "Mozilla/5.0"} # use headder to avoid 403-forbidden error
    response=requests.get("https://www.imdb.com/title/tt1187043/",headers=headers)
    soup=BeautifulSoup(response.text,"html.parser")
    #print(soup)
    movie=soup.find("main",class_="ipc-page-wrapper ipc-page-wrapper--base")
    #print(movie)
    for movie_data in movie:
        movie_name=movie_data.find("span",class_="hero__primary-text").text
        movie_rating=movie_data.find("div",class_="sc-4dc495c1-0 fUqjJu").span.text
        movie_year=movie_data.find("ul",class_="ipc-inline-list ipc-inline-list--show-dividers sc-cb6a22b2-2 aFhKV baseAlt baseAlt").li.a.text
        # movie_duration=movie_data.find("ul",class_="ipc-inline-list ipc-inline-list--show-dividers sc-cb6a22b2-2 aFhKV baseAlt baseAlt").li.text
        movie_gen=movie_data.find_all("a",class_="ipc-chip ipc-chip--on-baseAlt")
        movie_cast=movie_data.find_all("div",class_="sc-10bde568-5 dWhYSc")
        details = soup.find('ul', class_='ipc-inline-list ipc-inline-list--show-dividers sc-cb6a22b2-2 aFhKV baseAlt baseAlt')
        items = details.find_all('li')
        
        print("MOVIE NAME:",movie_name)
        print("RATING:",movie_rating)
        print("YEAR:",movie_year)
        
        for item in items:
            text = item.get_text(strip=True)
            if 'h' in text or 'm' in text:  # Looks for 'h' or 'm' to detect duration
                print("Duration:", text)
                break

            
        print("GENER:")
        for movie_genr in movie_gen:
            movie_gener=movie_genr.find("span",class_="ipc-chip__text").text
            print(movie_gener,end=", ")
        print()
        

        print("CAST:")    
        for movie_cas in movie_cast:
            cast=movie_cas.find("div",class_="sc-10bde568-7 bhMzVl").a.text
            print(cast,end=", ")

        

        
        #print("MOVIE_NAME:",movie_name)
        #print("RATING:",movie_rating)
        #print("YEAR:",movie_year)
        #print("DURATION:",movie_duration)
        #break
except Exception as e:
    print(e)


    
```
